# **Types of Linear Regression**

### 1. Simple Linear Regression

* One independent variable X and one dependent variable Y.

### 2. Multiple Linear Regression

* More than one independent variables ( predictors).

### 3. Polynomial Regression

* A type of linear regression where the relationships is modeled as a polynomial (non-linear in nature).

### 4. Ridge Regression

* A variant of linear regression that adds a penality (regularization) term to prevent overfitting.

### 5. Lasso Regression

* Another regularization version of linear regression, using __L1 penality__ to induce sparsity.

# **The Mathematics of Linear Regression**

      __Least Square Method__

$$
SSE = \sum \left( Y_i - \left( \beta_0 + \beta_1 X_i \right) \right)^2
$$

## Mathematics Behind Linear Regression

__y = mx + b__

y dependent variable ,x independent variable ,m slope ,b intercept.


**Objective:** Minimize SSE to find optimal values for \( \beta_0 \) and \( \beta_1 \).



### Normal Equations:

The slope coefficient:

$$
\beta_1 = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
$$

The intercept:

$$
\beta_0 = \bar{Y} - \beta_1 \bar{X}
$$

# **Multiple Linear Regression**

__Defination__ 

Multiple Linear Regression(MLR) is a statistical method used to predict an __Outcome(dependent variable)__ based on ___two or more input features (independent variable)__

__formula__

$$ 
Y = b_0 + b_1X_1 + b_2X_2 + ....+ b_nX_n + \varepsilon
$$

# **Evaluating Linear Regression Models**

Common metrics to assess model performance:

### **1. Mean Absolut Error(MAE)**

* Average of absolute defferences between predictions and actual values


$$
MAE = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

$$
y_i = actual.. value
$$

$$
hat{y}_i = predicted.. value
$$

$$
n= number of data points
$$


# **2. Mean Squared Error(MSE)**

* Average of squared differences between prediction and actal values.

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} \left( Y_i - \hat{Y}_i \right)^2
$$

### **3. Root Mean Squared Error(RMSE)**

* Square root of MSE, provides error in original units

$$
RMSE = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2 }
$$

### **4. R-squared**

* Propotion of variance explained by the model

# **Common Challenges in Linear Regression**

### 1. Multicollinearity

Highly correlated independent variables can distort coefficient estimates

### 2. Overfitting/Underfitting

Overfitting: Model fits training data too well but fails on test data.

### 3. Outliers

Extreme values can heavily influence the regression line and coefficients.

### 4.Non-Linearity

Assumes a linear relationship between features and target , fails if the relationship is non-linear.

### 5. Heteroscedasticity

Variance of residuals isn't constant,violating the assumption of homoscedasticity.

### 6. Autocorrelation

Residuals are correlated , common in time-series data

### 7. Feature Scalling

large defferences in feature scales can affect gradient descent convergence.

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
df = pd.read_csv("../datasets/data.csv")
# Display the first few rows of the dataset
print(df.head())
```

```python
# Display the summary of the dataset
df.describe()
```

```python
# Display the columns of the dataset
print("Columns in the dataset:")
print(df.columns)
print("Shape of the dataset:", df.shape)
print("Shape of the dataset:", df.shape)
print("Data types of the columns:")
print(df.dtypes)
print("check for missing values")
# Check for missing values
print(df.isnull().sum())
```

```python
# select numeric columns for regression
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("Numeric columns for regression:", numeric_columns)
```

```python
# remove numeric columns that are not suitable for regression
# For example, if 'id' is a column that should not be used in regression    
numeric_columns = numeric_columns.drop(['id', 'date', 'time'], errors='ignore')
print("Filtered numeric columns for regression:", numeric_columns)
```

```python
# Define feature X and target Y
X =df[['MALE (URBAN)']]
Y = df['FEMALE (URBAN)']  # Assuming '
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Create a Linear Regression model
model = LinearRegression()  
# Fit the model to the training data
model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)
```

```python
# plot regression line
plt.scatter(X_test, Y_test, color='blue', label='Actual Data')
plt.plot(X_test, Y_pred, color='red', linewidth=2, label='regression line')
plt.xlabel('Male (Urban)')
plt.ylabel('Female (Urban)')
plt.title('Linear Regression: Male (Urban) vs Female (Urban )')
plt.legend()
plt.show()
```

```python
# Display model coefficients
print("Model Coefficients:")
print(f'Intercept: {model.intercept_}')   
print(f'model.coef_: {model.coef_[0]}')  # Coefficients for the features
```
